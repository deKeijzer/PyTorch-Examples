# Creating a Convolutional Neural Network (CNN)
We are going to create a CNN that is able to process greyscale images of 32 by 32 pixels to perform classification on. We will classify between 10 different classes.  

To get a better understanding of convolutions see "A guide to convolutional arithmetic for deep learning": https://arxiv.org/abs/1603.07285

```python
import math
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as numpy

from tqdm import tqdm

class Net(nn.Module):

    def __init__(self):
        super(Net, self).__init__()
        # 1 input image channel, 6 output channels, 3x3 square convolution
        # kernel
        self.conv1 = nn.Conv2d(1, 6, 3)
        self.conv2 = nn.Conv2d(6, 16, 3)
        # an affine operation: y = Wx + b
        self.fc1 = nn.Linear(16 * 6 * 6, 120)  # 6*6 from image dimension
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        # Max pooling over a (2, 2) window
        x = F.max_pool2d(F.relu(self.conv1(x)), (2, 2))
        # If the size is a square you can only specify a single number
        x = F.max_pool2d(F.relu(self.conv2(x)), 2)
        x = x.view(-1, self.num_flat_features(x))
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x

    def num_flat_features(self, x):
        size = x.size()[1:]  # all dimensions except the batch dimension
        num_features = 1
        for s in size:
            num_features *= s
        return num_features


net = Net()
print(net)
```

```python
params = list(net.parameters())
print(len(params))
print(params[0].size())  # conv1's .weight
```

# Visualising weights
To get more insight into the neural network you can visualise the weights. Sometimes this helps in finding problems and e.g. explain why a model has stopped converging.

```python
# This cell has been generalised to work with different network architectures

weights = net.state_dict() # grab weights dict from the network

plt.figure(figsize=(10,len(weights)))


for i,layer in enumerate(weights): # for key in dict
    array = weights[layer].view(-1).numpy() # grab tensor, flatten, convert to numpy
    

    
    plt.subplot(math.ceil(len(weights)/2), 2, i+1) #nrows, ncols, plot index (+1 because plot index cannot be zero)
    plt.grid(alpha=0.5)
    plt.hist(array, ec="black") # draw black bin outlines
    plt.title(layer)
    plt.xlabel("Value [-]")
    plt.ylabel("Count [-]")
    plt.xticks(rotation=45)
    plt.tight_layout()


plt.show()
```

# Using the neural network
We can give the above neural network `Net()` an input and view the output. Lets work with some abstract art as an input image. We are going to feed the neural network a greyscale 'image' of 32 by 32 pixels. 

```python
image = torch.rand(1,1,32,32) # image, channel, width, height

_ = plt.imshow(image[0][0], cmap="gray") # plt takes (width, height) for single channel (greyscale) images
```

```python
output = net(image)
output
```

We have our predictions for the 10 different classes. If we were actually working with meaningful data, we would feed the model the ground truth $target$ also known as $\hat{y}$ to then backpropagate as follows.

## Training the model
If we where to train this network, the training would happen as follows.

```python
# Model input, X (input_) and the expected output, yhat (target)
input_ = image # Create a input_ variable for readability
target = torch.rand(1,10) # In this example we have a random ground truth


criterion = nn.MSELoss()
optimizer = optim.SGD(net.parameters(), lr=0.01) # define the optimizer


# Start out training loop
optimizer.zero_grad() # reset optimizer gradient
output = net(input_) # get model output, given input_

loss = criterion(output, target)
loss.backward() # backpropagate

optimizer.step() # Let the optimizer take a step (update weights)
```
