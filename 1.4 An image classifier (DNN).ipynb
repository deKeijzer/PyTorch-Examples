Train a simple Deep Neural Network (DNN) on the CIFAR-10 dataset. This dataset consists of images in 10 categories. There are 50.000 training images and 10.000 testing images. For more info on CIFAR see: https://www.cs.toronto.edu/~kriz/cifar.html

70% of the training data is used for training, 30% is used for validation. Validation takes place at the end of each epoch. Model weights are saved for each epoch. Once training is done the model weights from the desired epoch are loaded and the model is tested using a simple testing method.

# Notebook initialisation

```python
import random
import time
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import matplotlib.pyplot as plt

from tqdm import tqdm
from IPython.display import clear_output

%matplotlib inline
%config InlineBackend.print_figure_kwargs={'facecolor' : "w"} # Make sure the axis background of plots is white, this is usefull for t


# initialize random seeds
manualSeed = 1337
random.seed(manualSeed)
torch.manual_seed(manualSeed) 


# CUDA initialisation
ngpu = torch.cuda.device_count() # number of available gpus
device = torch.device("cuda:4") if (torch.cuda.is_available() and ngpu > 0) else "cpu" #cuda:0 for gpu 0, cuda:4 for gpu 5
torch.backends.cudnn.benchmark=True # Uses cudnn auto-tuner to find the best algorithm to use for your hardware

# CPU unitialisation
#device = "cpu"
#torch.set_num_threads(24) # number of threads for cpu algorithms
num_workers = 4 # number of processes for dataloaders
```

# Training parameters

```python
epochs = 20
batch_size = 64 # train & validation batch size
test_batch_size = 4

lr = 1e-3 # optimizer learning rate
```

# Loading data

```python
# normalise images
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

# load CIFAR-10 training dataset
dataset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)

# split training dataset into training and validation
trainset, valset = torch.utils.data.random_split(dataset, [int(len(dataset)*0.7), int(len(dataset)*0.3)])


# load CIFAR-10 test dataset
testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
```

# Set dataloaders
A dataloader loads data and feeds it to the neural network (net).
See PyTorch documentation for more information. 

```python
trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=num_workers, pin_memory=False)

valloader = torch.utils.data.DataLoader(valset, batch_size=batch_size,
                                          shuffle=True, num_workers=num_workers, pin_memory=False)

testloader = torch.utils.data.DataLoader(testset, batch_size=test_batch_size,
                                         shuffle=False, num_workers=num_workers, pin_memory=False)

classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

# Training loop
Adding the accuracy per epoch is left as an exercise for the reader.

```python
# loop parameters
save_weights_each_epoch = True
print_stats_every_n_batches = 250 # print stats every ... batches

net = net.to(device) # make sure the net is on device (so we can run this cell after we run the below cells)

# Initialize time variable and lists used to visualise training statistics
t0=time.time()
train_running_losses = [] # batch loss
val_running_losses = [] # batch loss
train_epoch_losses = [] # epoch loss
val_epoch_losses = [] # epoch loss

"""
Epoch loop
"""
for epoch in range(epochs):

    train_running_loss = 0.0
    val_running_loss = 0.0
    
    """
    Training loop
    """
    for i, data in enumerate(trainloader):
        t1 = time.time()
        inputs, labels = data
        actual_batch_size = len(inputs) # correct for dynamical batch size 
        inputs = inputs.to(device)
        labels = labels.to(device)

        # reset the parameter gradients to zero
        optimizer.zero_grad()

        
        outputs = net(inputs) # forward pass
        loss = criterion(outputs, labels)
        loss.backward() # backward pass
        optimizer.step() # optimize
        train_running_loss += (loss.item()/actual_batch_size) # calculate batch loss
        
        # print statistics
        if i % print_stats_every_n_batches == 0: #print stats every i % n batches
            print('Epoch [%d:%d] \t Train batch [%d:%d] \t train loss: %.4f \t time: %.3f [s]' %
                  (epoch + 1, epochs, (i + 1), len(trainloader), train_running_loss/((i+1)), t1-t0))
            t0 = time.time()
    


    """
    Validation loop
    """
    with torch.no_grad():
        for j, val_data in enumerate(valloader):
            t1 = time.time()
            inputs, labels = val_data
            actual_batch_size = len(inputs)

            inputs = inputs.to(device)
            labels = labels.to(device)

            outputs = net(inputs)
            val_loss = criterion(outputs, labels)

            val_running_loss += (val_loss.item()/actual_batch_size)
            
            # print stats
            if j % print_stats_every_n_batches == 0:
                print('Epoch [%d:%d] \t val batch [%d:%d] \t val loss: %.4f \t time: %.3f [s]' %
                      (epoch + 1, epochs, j + 1, len(valloader), val_running_loss/(j+1), t1-t0))
                # reset stats
                val_running_loss = 0.0
                t0 = time.time()

                
    """
    Epoch statistics
    """
    train_epoch_losses.append((train_running_loss/(len(trainloader)*batch_size))) # calculate mean epoch loss
    val_epoch_losses.append((val_running_loss/(len(valloader)*batch_size)))
    train_running_losses = []
    val_running_losses = []
    
    clear_output(wait=True)
    plt.plot(train_epoch_losses, ".-",label="Training loss")
    plt.plot(val_epoch_losses, "x-",label="Validation loss")
    plt.xlabel("Epoch [-]")
    plt.ylabel("Loss")
    plt.grid()
    plt.legend()
    plt.tight_layout()
    plt.show()
    
    if save_weights_each_epoch:
        PATH = "./data/weights/cifar_net_dnn_epoch_"+str(epoch)+".pth"
        torch.save(net.state_dict(), PATH)
        
        
print('Finished Training')
```
